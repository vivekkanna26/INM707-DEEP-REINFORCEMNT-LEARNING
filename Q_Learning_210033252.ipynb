## Step 3 :	Q LEARNING PARAMETERS AND POLICY 

#### Runing the algoriing sevelral times on the enviroments using the hyperparameter: 
1. Number of episodes = 300
2. Alpha ($\alpha$) learning rate = 0.9
2. Maximum number of steps per episode = 100. An episode ends when the agent reaches a goal state, or uses the maximum number of steps per episode
3. Gamma ($\gamma$) discount factor = 0.9
4. Epsilon ($\epsilon$) for $\epsilon$-greedy = 0.1 (10% of the time). 

```python
num_iters = 300
alpha = .9
gamma = .9
epsilon = 0.1
max_steps = 100
use_softmax_policy = False
env = MazeEnv()
q_hat, steps_vs_iters = qlearn(env, num_iters, alpha, gamma, epsilon, max_steps, use_softmax_policy)
plot_steps_vs_iters(steps_vs_iters)
plt.title(" Figure 1: Steps to goal vs episodes plot during training")
```

```python
# TODO: plot the policy from the Q values
plot_policy_from_q(q_hat, env)
```

## Step 5.1:	REPEAT THE EXPERIEMNTS WITH DIFFERENT PARAMETER AND POLCIES.
### POLICY - EPSILON

#### Inputting Epsilon value is changes to [.01] ,[0.5],[0.9]

#### We know that increasing the epsilon values singlicately make thbe algorithm greedy and makes the policy learn quicker by not so viable in the long run.Lets us expiement the same


```python

num_iters = 300
alpha = 0.9
gamma = .9
epsilon = .1
max_steps = 100
use_softmax_policy = False


epsilon_list = [.01,.5,.9]

env = MazeEnv()

steps_vs_iters_list = []
for epsilon in epsilon_list:
    q_hat, steps_vs_iters = qlearn(env, num_iters, alpha, gamma, epsilon, max_steps, use_softmax_policy)
    steps_vs_iters_list.append(steps_vs_iters)
```

```python

label_list = ["epsilon={}".format(eps) for eps in epsilon_list]
plot_several_steps_vs_iters(steps_vs_iters_list, label_list)
plt.title(" Figure 3:  Steps to goal vs episodes plot ")
```

## Step 5.2:	REPEAT THE EXPERIEMNTS WITH DIFFERENT PARAMETER AND POLCIES.
### Learning rate Alpha 

### Alpha value is changed to [0.01] ,[0.5],[0.8]

### The higher amount of alpha signifies the learning can be ouccer at a faster phase


```python

num_iters = 300
alpha = 0.9
gamma = .9
epsilon = .1
max_steps = 100
use_softmax_policy = False

alpha_list = [.01,.5,.8]

env = MazeEnv()

steps_vs_iters_list = []
for alpha in alpha_list:
    q_hat, steps_vs_iters = qlearn(env, num_iters, alpha, gamma, epsilon, max_steps, use_softmax_policy)
    steps_vs_iters_list.append(steps_vs_iters)
```

```python

label_list = ["alpha={}".format(eps) for eps in alpha_list]
plot_several_steps_vs_iters(steps_vs_iters_list, label_list)
plt.title("figure 4 : Steps to goal vs episodes plot ")
```

## Step 5.2: REPEAT THE EXPERIEMNTS WITH DIFFERENT PARAMETER AND POLCIES.
### Discount factor Gamma

###  Gamma value is changed to [.01] ,[0.4],[0.8]
### Seting the gamma to lower amount will make the reward less considering the current rewards and gamma value higher will consider the rewards to be significantly good in the longer term

```python

num_iters = 300
alpha = 0.9
gamma = .9
epsilon = .1
max_steps = 100
use_softmax_policy = False


gamma_list = [.01,.4,.8]

env = MazeEnv()

steps_vs_iters_list = []
for gamma in gamma_list:
    q_hat, steps_vs_iters = qlearn(env, num_iters, alpha, gamma, epsilon, max_steps, use_softmax_policy)
    steps_vs_iters_list.append(steps_vs_iters)
```

```python

label_list = ["gamma={}".format(eps) for eps in alpha_list]
plot_several_steps_vs_iters(steps_vs_iters_list, label_list)
plt.title("Figure 5: Steps to goal vs episodes plot")
```
